```python
# Convert Color image to GrayScale


## Reading the image
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('./Images/Obama.jpg')
cv2.imshow('Obama_Color', image)
cv2.waitKey()
cv2.destroyAllWindows()

```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
## Conversion to Grayscale

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Gray Scale Image Obama',gray_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(gray_image,  cmap ='gray')
```

```python
# Directly loading the gray scale images. 

# pass ->0 as parameter
image_gray = cv2.imread('./Images/Obama.jpg', 0)
cv2.imshow('Gray Scale Image', image_gray)
cv2.waitKey()
cv2.destroyAllWindows()
```
